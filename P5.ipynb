```python
import pandas as pd
import numpy as np
import tensorflow as tf
```

```python
CSV_COLUMN_NAMES = ["SepalLength", "SepalWidth", "PetalLength", "PetalWidth", "Species"]
SPECIES = ["Setosa", "Versicolor", "Virginica"]
train = pd.read_csv("sample_data/iris_training.csv", names = CSV_COLUMN_NAMES, header=0)
test = pd.read_csv("sample_data/iris_test.csv", names = CSV_COLUMN_NAMES, header=0)
```

```python
train.head(5)
```

```python
test.head(5)
```

```python
train_y = train.pop("Species")
test_y = test.pop("Species")
```

```python
train.head()
```

```python
train.shape
```

```python
feature_columns = []
for key in train.keys():
    feature_columns.append(tf.feature_column.numeric_column(key=key))
print(feature_columns)
```

```python
from tensorflow.keras import models, layers
model = models.Sequential([
    layers.Dense(30, activation='relu', input_shape=(4,)),
    layers.Dense(10, activation='relu'),
    layers.Dense(3, activation='softmax')  # 3 classes
])
```

```python
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```python
model.fit(train, train_y, epochs=50, validation_split=0.2)
```

```python
test_loss, test_accuracy = model.evaluate(test, test_y)
print(f"Test accuracy: {test_accuracy:.4f}")
```

```python
predictions = model.predict(test)
predicted_classes = np.argmax(predictions, axis=1)
print(predicted_classes[:5])
```

```python
features = ["SepalLength", "SepalWidth", "PetalLength", "PetalWidth"]
user_input = []

print("Please type numeric values as prompted: ")
for feature in features:
    while True:
        val = input(f"{feature}: ")
        try:
            user_input.append(float(val))
            break
        except ValueError:
            print("Please enter a valid number.")

user_input_array = np.array([user_input])

prediction_probs = model.predict(user_input_array)[0]
predicted_class = np.argmax(prediction_probs)
predicted_species = SPECIES[predicted_class]
predicted_prob = prediction_probs[predicted_class]

print(f"Predicted species: {predicted_species} ({predicted_prob * 100:.2f}%)")
```
