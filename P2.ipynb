```python
import pandas as pd
import numpy as np

df = pd.read_csv('online_sales_dataset.csv') 
df.head()
```

```python
df.dtypes
```

```python
df["InvoiceDate"] = pd.to_datetime(df['InvoiceDate'])
df["InvoiceDate"]
df.dtypes
```

```python
df["Total"] = df["Quantity"] * df["UnitPrice"] * (1 - df["Discount"])
df.head()
```

```python
df["Date"] = df["InvoiceDate"].dt.date
df.head()
```

```python
weekday_order = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday","Sunday"]
df["Weekday"] = pd.Categorical(
    df["InvoiceDate"].dt.day_name(),
    categories =  weekday_order,
    ordered = True
)
df.head()
```

```python
daily_sales = df.groupby('Date')["Total"].sum().reset_index()
daily_sales.head()
```

```python
daily_sales["Rolling7DayAvg"] = daily_sales["Total"].rolling(window = 7,
                                                             min_periods = 7).mean()
daily_sales.head(21)
```

```python
df['Date'] = pd.to_datetime(df["Date"])
df['Date']
```

```python
monthly_sales = df.set_index("Date")["Total"].resample("ME").sum()
monthly_sales
```

```python
monthly_sales_growth = monthly_sales.pct_change() * 100
monthly_sales_growth
```

```python
def calculate_profit_margin(row):
    if pd.isna(row["Total"]) or row["Total"] == 0:
        return 0
    return (row["Total"] - row["ShippingCost"]) / row["Total"] * 100

df["ProfitMargin"] = df.apply(calculate_profit_margin, axis=1)
lowest_margin = df["ProfitMargin"].min()
print(lowest_margin)
df.loc[df["ProfitMargin"].idxmin()]
```

```python
df.head()
```

```python
country_analysis = df.groupby('Country').agg({
    'Total': ['sum', 'mean', 'count'],
    'Discount': ['mean', lambda x:(x>0.3).sum()],
    'ProfitMargin': ['mean','median'],
    'CustomerID':pd.Series.nunique
}).round(2)
country_analysis
```

```python
sales_pivot = pd.pivot_table(
    df,
    values = ["Total", "Quantity"],
    index=["Country","Category"],
    columns=["Weekday"],
    aggfunc={"Total":'sum',
             "Quantity":'mean'},
    fill_value = 0
).sort_index(level=[0,1])
sales_pivot.head()
```

```python
channel_category = pd.crosstab(
    df["SalesChannel"],
    df["Category"],
    values = df["Total"],
    aggfunc = 'sum',
    normalize = 'index'
)*100
channel_category
```

```python
product_metrics = df.groupby('Description').agg({
    "Quantity":"sum",
    "Total":"sum",
    "InvoiceNo":"count"
}).assign(
    AvgOrderValue = lambda x:x['Total']/x["InvoiceNo"],
    UnitsPerOrder = lambda x:x["Quantity"]/x["InvoiceNo"]
).round(2)
product_metrics
```

```python
high_value_transactions = df.query("Total > Total.mean() and Discount <0.2 and ReturnStatus == 'Not Returned' ")
high_value_transactions
```

```python
category_performance = df.groupby('Category').agg({
    "Total":["sum","mean","count"],
    "Discount":"mean",
    "ReturnStatus": lambda x:(x=="Returned").mean()*100
})
category_performance.columns = ["TotalSales", "AvgSales", "NumOrders", "AvgDiscount", "ReturnRate"]
category_performance
```
